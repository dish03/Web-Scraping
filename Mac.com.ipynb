### Importing modules`

```python
import requests
import re
from bs4 import BeautifulSoup
from lxml import etree
import pandas as pd
import numpy as np
import matplotlib as mp
import pprint
```

### Passing URL to BS4

```python
URL = "https://www.maccosmetics.com/collections"
r = requests.get(URL)

soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
# print(soup.prettify())
```

### Extracting image URL and product name

```python
dict_list = soup.find_all('div', attrs = {'class': "h-full pt-5 pb-8 bg-product-card-bg"}) 
value = []
for i in dict_list:
  # print(i)
  dic = {}
  dic['image_url']=i.source['srcset']
  dic['product_name'] = i.h2.text.strip()
  value.append(dic)

pprint.pprint(value)
```

```python
image_url_list = []
product_name_list = []

for i in value:
  image_url_list.append(i['image_url'])
  product_name_list.append(i['product_name'])

print(image_url_list)
print(product_name_list)
```

### Fetching original product prize

```python
original_product_prize = soup.findAll('span',attrs={'class':'font-bold'})
original_prize_list=[]
for i in original_product_prize:
    prize = i.text.strip()
    original_prize_list.append(prize)
print(original_prize_list)
```

### Fetching discounted prize for few products

```python
discounted_product_prize = soup.findAll('span',attrs={'class':'ml-2'})
discounted_prize_list=[]
for i in discounted_product_prize:
    prize = i.text.strip()
    discounted_prize_list.append(prize)
print(discounted_prize_list)
```

### Fetching product feature

```python
product_feature = soup.findAll('p',attrs={'class':'font-headline text-center mx-auto max-w-[380px] px-5'})
product_feature_list=[]
for i in product_feature:
  feature = i.text.strip()
  product_feature_list.append(feature)
print(product_feature_list)
```

### Extracting URLs for individual products

```python
product_url = soup.findAll('div',attrs={'class':'h-full pt-5 pb-8 bg-product-card-bg'})
url_list = []
# print(product_url)
for i in product_url:
    a_tag = i.find('a')
    url = a_tag['href']
    # print(url)
    url_list.append(url)
print(url_list)
```

### Extracting product benefits and ingredients

```python
from bs4 import BeautifulSoup
import requests

base_url = "https://www.maccosmetics.com"
benefit_list = []
ingredient_list = []

for url in url_list:
    full_url = base_url + url
    response = requests.get(full_url)
    soup = BeautifulSoup(response.content, 'html5lib')
    
    benefit_value = soup.find('div', attrs={'id': 'product-overview--about'})
    ingredient_value = soup.find('div', attrs={'id': 'product-overview--ingredients'})
    
    
    
    if benefit_value:
        benefit_text = benefit_value.get_text().strip()
        benefit_list.append(benefit_text)
    else:
        benefit_list.append("NA")
        
    if ingredient_value:
        ingredient_text = ingredient_value.get_text().strip()
        ingredient_list.append(ingredient_text)
    else:
        ingredient_list.append("NA")
        
print(benefit_list)
print(ingredient_list)
```

### Extracting product usage

```python
from bs4 import BeautifulSoup
import requests

base_url = "https://www.maccosmetics.com"

usage_list = []

for url in url_list:
    full_url = base_url + url
    response = requests.get(full_url)
    soup = BeautifulSoup(response.content, 'html5lib')
    
    usage_value = soup.find('div', attrs={'id': 'product-overview--product_usage'})
    
    if usage_value:
        usage_text = usage_value.get_text().strip()
        usage_list.append(usage_text)
    else:
        usage_list.append("Not found")
print(usage_list)
```

### Preprocessing and Removing unnecessary words

```python
def clean_ingredients(ingredient):
    cleaned_ingredient = re.sub(r'\([\w\s]+\)', '', ingredient)  # Remove codes
    cleaned_ingredient = cleaned_ingredient.strip()  # Remove leading/trailing space
    cleaned_ingredient = re.sub(r',\s*\d+', '', cleaned_ingredient)  # Remove numeric values
    cleaned_ingredient = cleaned_ingredient.replace('/', '')  # Replace forward slashes with spaces
    cleaned_ingredient = cleaned_ingredient.replace(']','')
    cleaned_ingredient = cleaned_ingredient.replace('[+ -','')  # Remove '[+/-'
    cleaned_ingredient = re.sub(r'PLEASE BE AWARE THAT.*$', '', cleaned_ingredient)  # Remove disclaimer
    return cleaned_ingredient

# Apply the function to the ingredients column
df['Cleaned Ingredients'] = df['Ingredients'].apply(clean_ingredients)
```

```python
df.head()
```

### Creating a bag of words (corpus of all ingredients)

```python
ingredient_idx = {}
corpus = []
idx = 0 

for i in range(len(df)):    
    ingredients = df['Cleaned Ingredients'][i]
    ingredients_lower = ingredients.lower()
    tokens = ingredients_lower.split(', ')
    corpus.append(tokens)
    for ingredient in tokens:
        if ingredient not in ingredient_idx:
            ingredient_idx[ingredient] = idx
            idx += 1
```

```python
print(ingredient_idx)
```

### Creating a Document-Term matrix

```python
M = len(df)
N = len(ingredient_idx)

A = np.zeros((M,N))
```

```python
# Define the oh_encoder function
def oh_encoder(tokens):
    x = np.zeros(N)    
    for ingredient in tokens:
        # Get the index for each ingredient
        idx = ingredient_idx[ingredient]
        # Put 1 at the corresponding indices
        x[idx] = 1
    return x
```

```python
# Make a document-term matrix
i = 0
for tokens in corpus:
    A[i, :] = oh_encoder(tokens)
    i += 1
```
